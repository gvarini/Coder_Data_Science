# MODELS

### LINEAL REGRESSION WITH MNT MEAT PRODUCTS

```python
df_meat_consumers.shape
```

```python
df_meat_consumers_target = df_meat_consumers[['Income', 'MntMeatProducts']]

# Definimos variable predictora y variable target
X = df_meat_consumers_target.drop(['MntMeatProducts'], axis=1)
y = df_meat_consumers_target['MntMeatProducts']

# Separamos datos en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Creamos modelo de regresión lineal
model = LinearRegression()

# Entrenamos el modelo con conjunto de entrenamiento
model.fit(X_train, y_train)

# Generamos las predicciones
preds_train = model.predict(X_train)
preds_test = model.predict(X_test)

# Calculamos los coeficientes beta
print('COEFFICIENTS')
print(model.coef_)
print(model.intercept_)

# Evaluamos los modelos
print('EVALUATION')
mse_test = mean_squared_error(y_test, preds_test)
mse_train = mean_squared_error(y_train, preds_train)
print('MSE - Test: {:.2f}, Train: {:.2f}'.format(mse_test, mse_train))

# Calculate the evaluation metrics
mae_test = mean_absolute_error(y_test, preds_test)
rmse_test = np.sqrt(mse_test)
score = model.score(X_test, y_test)

# Print the evaluation metrics
print('Mean Absolute Error (MAE) - Test: {:.3f}'.format(mae_test))
print('Root Mean Squared Error (RMSE) - Test: {:.3f}'.format(rmse_test))
print('R-squared (R2) Score: {:.3f}'.format(score))
```

```python
plt.scatter(X_test['Income'], y_test)
plt.plot(X_test['Income'], preds_test, color='red', linewidth=2)
plt.xlabel('Income')
plt.ylabel('MntMeatProducts')
plt.show()
```

### Bias-Variance

```python
# List to store the evaluation metrics
degrees = []
train_mse = []
test_mse = []

# Try different polynomial degrees from 1 to 9
for degree in range(1, 10):
    # Create polynomial features
    poly = PolynomialFeatures(degree=degree)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)

    # Create polynomial regression model
    model = LinearRegression()
    model.fit(X_train_poly, y_train)

    # Generate the predictions
    preds_train = model.predict(X_train_poly)
    preds_test = model.predict(X_test_poly)

    # Calculate the evaluation metrics (MSE)
    train_mse.append(mean_squared_error(y_train, preds_train))
    test_mse.append(mean_squared_error(y_test, preds_test))
    degrees.append(degree)

# Plot the evaluation metrics against polynomial degree
plt.plot(degrees, train_mse, label='Train MSE')
plt.plot(degrees, test_mse, label='Test MSE')
plt.xlabel('Polynomial Degree')
plt.ylabel('Mean Squared Error (MSE)')
plt.legend()

# Add values to the plot
for i, degree in enumerate(degrees):
    plt.text(degree, train_mse[i], f'{train_mse[i]:.2f}', ha='center', va='bottom')
    plt.text(degree, test_mse[i], f'{test_mse[i]:.2f}', ha='center', va='bottom')

plt.gcf().set_size_inches(12, 6)

plt.show()
```
