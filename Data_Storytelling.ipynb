## What are the marital states that consume more meat and its percentage?





```python
# Create a figure with two subplots
fig, axs = plt.subplots(ncols=3, figsize=(16, 6))

# Total MntMeatProducts by marital status
meat_by_marital = df_no_outliers.groupby('Marital_Status')['MntMeatProducts'].sum()
sns.barplot(x=meat_by_marital.index, y=meat_by_marital, ax=axs[0])
axs[0].set_title('Total MntMeatProducts consumed by marital status')
axs[0].set_xlabel('Marital Status')
axs[0].set_ylabel('Total MntMeatProducts')

# Add the value labels to the first subplot
for i, v in enumerate(meat_by_marital):
    axs[0].text(i, v + 1000, f'{v:,}', horizontalalignment='center', color='black')

df_meat_marital = df_no_outliers[['Marital_Status', 'MntMeatProducts']]

# Group the dataset by marital status and sum the MntMeatProducts column for each group
meat_by_marital = df_meat_marital.groupby('Marital_Status')['MntMeatProducts'].sum()

# Calculate the total number of meat consumers across all marital statuses
total_meat_consumers = meat_by_marital.sum()

# Calculate the percentage of meat consumers for each marital status
percentages = meat_by_marital / total_meat_consumers * 100

# Create a horizontal bar plot to show the percentages
sns.barplot(x=percentages, y=percentages.index, ax=axs[1])
axs[1].set_xlabel('Percentage of meat consumers')
axs[1].set_ylabel('Marital Status')
axs[1].set_title('Percentage of meat consumers comparing marital status')


# Percentage of meat consumers by marital status
meat_consumers = df_meat_marital[df_meat_marital['MntMeatProducts'] > 0].groupby('Marital_Status')['MntMeatProducts'].count() / df_meat_marital.groupby('Marital_Status')['MntMeatProducts'].count() * 100

sns.barplot(x=meat_consumers.index, y=meat_consumers, ax=axs[2])
axs[2].set_title('Percentage of meat consumers by marital status')
axs[2].set_xlabel('Marital Status')
axs[2].set_ylabel('Percentage of meat consumers')

# Add the value labels to the second subplot
for i, v in enumerate(meat_consumers):
    axs[2].text(i, v + 1, f'{v:.1f}%', horizontalalignment='center', color='black')

plt.tight_layout()
plt.show()
```

These subplots provide a comprehensive view of meat consumption patterns based on marital status. They allow us to compare the total consumption, overall distribution, and relative proportion of meat consumers among different marital status categories.

Total MntMeatProducts consumed by marital status helps compare the overall consumption of meat products among different marital status groups.

Percentage of meat consumers comparing marital status provides insights into the distribution of meat consumers across different marital status groups.

Percentage of meat consumers by marital status focuses specifically on the percentage of individuals within each marital status group who consume meat products.

# LINEAL REGRESSION FOR INCOME AND MNT MEAT PRODUCTS

```python
df_meat_consumers.shape
```

```python
df_meat_consumers_target = df_meat_consumers[['Income', 'MntMeatProducts']]

# Definimos variable predictora y variable target

X = df_meat_consumers_target.drop(['MntMeatProducts'], axis=1)
y = df_meat_consumers_target['MntMeatProducts']

# Separamos datos en conjunto de entrenamiento y conjunto de prueba

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Creamos modelo de regresión lineal

model = LinearRegression()

# Entrenamos el modelo con conjunto de entrenamiento

model.fit(X_train, y_train)

# Generamos las predicciones

preds_train = model.predict(X_train)
preds_test = model.predict(X_test)

# Calculamos los caeficientes beta

print('COEFFICIENTS')
print(model.coef_)
print(model.intercept_)

# Evaluamos los modelos

print('EVALUATION')
mse_test = mean_squared_error(y_test, preds_test)
mse_train = mean_squared_error(y_train, preds_train)
print('MSE - Test: {:.2f}, Train: {:.2f}'.format(mse_test, mse_train))
# calculamos el coeficiente de determinacion

score = model.score(X_test, y_test)
print('R-squared value: {:.2f}'.format(score))
```

```python
plt.scatter(X_test['Income'], y_test)
plt.plot(X_test['Income'], preds_test, color='red', linewidth=2)
plt.xlabel('Income')
plt.ylabel('MntMeatProducts')
plt.show()
```

**Conclusions**

The R-squared value is 0.50, which means that approximately 50% of the variance in the amount spent on meat products can be explained by the income variable. This indicates a moderate level of predictability, suggesting that income alone may not be the sole factor influencing meat product consumption.

Considering these results, the linear regression model provides some insights into the relationship between income and meat product consumption. However, it's important to note that the model's performance can be further improved, as there is still a significant amount of unexplained variance in the data (as indicated by the MSE and R-squared values).

### WHAT IF WE ADD NUM STORE PURCHASES

```python
fig = go.Figure( go.Scatter3d(x=df_meat_consumers['Income'],
                              y=df_meat_consumers['MntMeatProducts'],
                              z=df_meat_consumers['NumStorePurchases'],
                              mode='markers'))
fig.show()
```

```python
# Create a subset of the data with only the columns we need

data = df_meat_consumers[['Income', 'MntMeatProducts', 'NumStorePurchases']]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data[['Income', 'NumStorePurchases']], data['MntMeatProducts'], test_size=0.3, random_state=1)

# Create the linear regression model and fit it to the training data
model = LinearRegression()
model.fit(X_train, y_train)

# Use the model to make predictions on the testing data
predictions = model.predict(X_test)

# Add the predicted values to the test data frame
test_data = X_test.copy()
test_data['MntMeatProducts'] = y_test
test_data['Predicted'] = predictions

# Calculate the R-squared value to evaluate the model's performance
r_squared = model.score(test_data[['Income', 'NumStorePurchases']], test_data['MntMeatProducts'])
print(f"R-squared value: {r_squared}")

# Calculate the mean squared error
mse = mean_squared_error(test_data['MntMeatProducts'], test_data['Predicted'])
print(f"MSE (Mean squared error): {mse}")

# Calculate the mean absolute percentage error
mape = mean_absolute_percentage_error(test_data['MntMeatProducts'], test_data['Predicted'])
print(f"MAPE (mean absolute percentage error): {mape}")

# Create a scatter plot with real and predicted values
fig = px.scatter(test_data, x='Income', y='MntMeatProducts', color='NumStorePurchases', trendline='ols', title='Real and Predicted MntMeatProducts')
fig.add_scatter(x=test_data['Income'], y=test_data['Predicted'], mode='markers', name='Predicted')
fig.show()
```

By adding the variable 'NumStorePurchases', we achieved a higher R-squared value of 0.5424, indicating that it explains a larger proportion of the variance in meat product consumption compared to the first model's R-squared value of 0.50. Additionally, has a lower mean squared error (MSE) of 9165.63, indicating smaller prediction errors compared to the first model's MSE of 9992.53.

**Conclusion**

In conclusion, the second model, which incorporates both income and the number of store purchases as predictors, demonstrates superior performance in predicting meat product consumption compared to the first model that solely relied on income. The second model exhibits a higher R-squared value, lower MSE, and lower MAPE, indicating a better fit to the data and more accurate predictions.
